## PDF and CDF

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

```python
x = np.linspace(-6, 6, 1000)
```

```python
fx = norm.pdf(x, loc=0.0, scale=1.0)
```

```python
plt.plot(x, fx);
```

```python
Fx = norm.cdf(x, loc=0.0, scale=1.0)
```

```python
plt.plot(x, Fx);
```

```python
logfx = norm.logpdf(x, loc=0.0, scale=1.0)
```

```python
plt.plot(x, logfx);
```

## Convolution

```python
from PIL import Image
```

```python
im = Image.open('lena.png')
```

```python
gray = np.mean(im, axis=2)
```

```python
x = np.linspace(-6, 6, 50)
```

```python
fx = norm.pdf(x, loc=0.0, scale=1.0)
```

```python
filt = np.outer(fx, fx)
```

```python
plt.imshow(filt, cmap='gray');
```

```python
from scipy.signal import convolve2d
```

```python
out = convolve2d(gray, filt)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(out, cmap='gray');
```
